# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
# The first graph presents the data clearly visualizing the concurrent passing of time in chronological order (week 1 data then week 2 data)
# It also helps to visualize the trend with respective to normalizing the number of visitors having different y axis values
# the graph does take some initial calibration given the y axis is in the middle.
# The second graph help to visualize the data directly and is much easier to understand the representation of the data as they are using the same axis figures
# one disadvantage is you don't know which week came before the other, so there may impact of the timing of visitors coming more in a prior week or vice a versa
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014]

prices_A = [430, 450, 460, 450, 400, 300, 270]
prices_B = [410, 430, 440, 430, 380, 290, 260]
prices_C = [0, 0, 100, 180, 250, 230, 190]
prices_D = [0, 0, 90, 160, 200, 210, 220]
prices_E = [0, 0, 0, 100, 180, 220, 250]

plt.plot(years, prices_A, label='A')
plt.plot(years, prices_B, label='B')
plt.plot(years, prices_C, label='C')
plt.plot(years, prices_D, label='D')
plt.plot(years, prices_E, label='E')

plt.xlabel("Year")
plt.ylabel("Average price")
plt.title("Retail price over time")

plt.yticks(range(0, 600, 100), [f'${v}' for v in range(0, 600, 100)])

plt.text(2014.1, prices_A[-1], 'A', va='center')
plt.text(2014.1, prices_B[-1], 'B', va='center')
plt.text(2014.1, prices_C[-1], 'C', va='center')
plt.text(2014.1, prices_D[-1], 'D', va='center')
plt.text(2014.1, prices_E[-1], 'E', va='center')

plt.tight_layout()
plt.show()
```
