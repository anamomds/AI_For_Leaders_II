```python
diabetes_survey = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
diabetes_survey.iloc[0:5]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

diabetes_survey["BMI"].plot.hist()
plt.title('Histogram of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
diabetes_survey["BMI"].plot.box()
plt.title('Box Plot of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.violinplot(x='BMI', data=diabetes_survey)
plt.title('Violin Plot of BMI')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.swarmplot(x='BMI', data=diabetes_survey)
plt.title('Swarm Plot of BMI')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Ride A', 'Ride B', 'Ride C']
series1 = [10, 15, 7]
series2 = [12, 9, 11]

y = np.arange(len(categories))
bar_height = 0.35

plt.figure(figsize=(8, 4))
plt.barh(y - bar_height/2, series1, height=bar_height, label='Adults', color='skyblue')
plt.barh(y + bar_height/2, series2, height=bar_height, label='Children', color='salmon')

plt.yticks(y, categories)
plt.xlabel('Number of Riders')
plt.title('Ride Participation by Group')
plt.legend()
plt.tight_layout()
plt.show()
```
